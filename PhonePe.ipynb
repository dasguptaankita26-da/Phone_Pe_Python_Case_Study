Task 3: Data Quality Checks
3.1: Ensure data consistency across state and district levels

1. For each state, calculate the total number of transactions, total transaction
amount, and total registered users by summing up the values from the district
level data.


```python
a=District_Txn_df.groupby('State').agg({'Transactions':'sum','Amount (INR)':'sum','Registered Users':'sum'}).reset_index()
a.columns=['State',	'No of Transactions','Total Amount','Total Registered Users']
a
```

2. Compare the results with the corresponding values at the state level to ensure
they match.

```python
b= State_Txn_df.groupby('State').agg(
    {'Transactions': 'sum',
     'Amount (INR)': 'sum',
     'Registered Users': 'sum'}
).reset_index()
b.columns = ['State', 'No of Transactions', 'Total Amount', 'Total Registered Users']
b
```

**Comparison**

```python
comparison = a.merge(b, on='State', suffixes=('_District', '_State'))

comparison.head(5)


comparison['Transactions_Match'] = comparison['No of Transactions_District'] == comparison['No of Transactions_State']
comparison['Amount_Match'] = comparison['Total Amount_District'] == comparison['Total Amount_State']
comparison['Users_Match'] = comparison['Total Registered Users_District'] == comparison['Total Registered Users_State']

print("\nMatch Results:")
print(comparison[['State', 'Transactions_Match', 'Amount_Match', 'Users_Match']])
```

3. Display any discrepancies found between the district-level and state-level data.

```python
# Find discrepancies
discrepancies = comparison[
    (comparison['Transactions_Match'] == False) |
    (comparison['Amount_Match'] == False) |
    (comparison['Users_Match'] == False)
]

# Display
if len(discrepancies) == 0:
    print(" No discrepancies found!")
else:
    print(f" {len(discrepancies)} states have discrepancies:")
    print(discrepancies)
```

6.1: Identify any trends or patterns in the transaction data

**PhonePe transactions show consistent and exponential growth quarter over quarter from 2018 to 2021, with total transactions increasing nearly 30x from ₹0.1B in Q1 2018 to ₹4.0B in Q2 2021.**

```python
#Total Transaction Over year
a=State_Txn_df.groupby(['Year','Quarter']).agg({'Transactions':'sum'}).reset_index()
a['Year-Q']=a['Year'].astype('str')+'-Q'+a['Quarter'].astype('str')
a.columns=['Year','Quarter','Total Transaction','Year-Q']
sns.barplot(data=a,x='Year-Q',y='Total Transaction')
plt.title("Total Transaction Over year")
plt.xticks(rotation=90)
plt.show()
```

 **Peer-to-peer payments show the highest and fastest growth over time, becoming the dominant transaction type by 2021.**

```python
# Trend 2: Transaction Type distribution over years
a = State_TxnSplit_df.groupby(['Year', 'Quarter', 'Transaction Type']).agg(
    {'Transactions': 'sum'}
).reset_index()

a.columns = ['Year', 'Quarter', 'Transaction Type', 'Total_Transactions']  # ✅ rename first

# ✅ Create Year-Quarter column after renaming
a['Year_Quarter'] = a['Year'].astype(str) + '-Q' + a['Quarter'].astype(str)

plt.figure(figsize=(12, 5))
sns.lineplot(data=a, x='Year_Quarter', y='Total_Transactions', hue='Transaction Type', marker='o')
plt.title('Transaction Type Trends Over Years')
plt.xlabel('Year - Quarter')
plt.ylabel('Total Transactions')
plt.xticks(rotation=90)
plt.show()
```

**Transactions are growing faster than users → higher engagement per user**

```python
#Growth of Registered Users vs Transactions
a = State_Txn_df.groupby(['Year','Quarter']).agg({
    'Registered Users':'sum',
    'Transactions':'sum'
}).reset_index()

a['Year-Q'] = a['Year'].astype(str) + '-Q' + a['Quarter'].astype(str)

sns.lineplot(data=a, x='Year-Q', y='Registered Users', label='Users')
sns.lineplot(data=a, x='Year-Q', y='Transactions', label='Transactions')

plt.xticks(rotation=90)
plt.title('User Growth vs Transaction Growth')
plt.show()
```

**ATV increasing → users making higher-value transactions**

```python
#Average Transaction Value (ATV) Trend
a = State_Txn_df.groupby(['Year','Quarter'])['ATV (INR)'].mean().reset_index()
a['Year-Q'] = a['Year'].astype(str) + '-Q' + a['Quarter'].astype(str)

sns.lineplot(data=a, x='Year-Q', y='ATV (INR)')
plt.xticks(rotation=90)
plt.title('Average Transaction Value Trend')
plt.show()
```

**Karnataka has the highest transaction volume, followed closely by Maharashtra, indicating these are the most active digital payment states.**

 Correlate demographic data with transaction data

1. Find correlations between demographic data (e.g., population density) and
transaction data (e.g., transaction volume). Summarize your findings. [Hint: you
can use corr()]


**Transaction volume is strongly driven by user engagement and platform activity rather than demographic factors. While transactions and total amount show a near-perfect relationship, registered users and app opens have a strong positive impact on transaction growth. Population and density show only moderate influence, indicating that digital adoption depends more on user behavior than sheer population size.**

```python
# Step 1: Merge datasets
merged = District_Txn_df.merge(
    District_demo_df,
    on=['State', 'District']
)


# Step 2: Select relevant numeric columns
cols = [
    'Population',
    'Density',
    'Transactions',
    'Amount (INR)',
    'Registered Users',
    'App Opens'
]

# Step 3: Correlation matrix
corr_matrix = merged[cols].corr()

print("Correlation Matrix:\n")
print(corr_matrix)

# Step 4: Heatmap visualization
plt.figure(figsize=(4,3))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation between Demographics and Transactions')
plt.show()
```
